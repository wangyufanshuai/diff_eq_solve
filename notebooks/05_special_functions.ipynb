This notebook covers 12 classical ordinary differential equations whose solutions define the **special functions of mathematical physics**. These functions arise repeatedly in problems involving the Laplacian in curvilinear coordinates, quantum mechanics, electromagnetic theory, and statistical mechanics.

Each section presents:
1. The defining ODE and its symbolic solution
2. A numerical solution plotted via the `plot_special_function` visualizer

All equations are registered in the `diff_eq_solver` equation registry.

## 1. Bessel Equation

The Bessel equation of order $n$ is:

$$x^2 y'' + x y' + (x^2 - n^2) y = 0$$

Solutions are Bessel functions of the first kind $J_n(x)$ and second kind $Y_n(x)$.

**Applications:** Cylindrical wave propagation, heat conduction in cylinders, diffraction patterns, FM signal synthesis.

```python
# Symbolic solution
eq = registry.get('bessel_equation')
print("Bessel Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot J_0, J_1, J_2 on same axes
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (0.01, 20)
x_eval = np.linspace(0.01, 20, 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
labels = ['$J_0(x)$', '$J_1(x)$', '$J_2(x)$']

for n_val, color, label in zip([0, 1, 2], colors, labels):
    # J_n(0) = delta_{n,0}, J_n'(0) from recurrence
    if n_val == 0:
        y0 = [1.0, 0.0]
    elif n_val == 1:
        y0 = [0.0, 0.5]
    else:
        y0 = [0.0, 0.0]
    result = eq.numerical_solve(x_span, y0, t_eval=x_eval, params={'n': n_val})
    plot_special_function(result, ax=ax, label=label, color=color)

ax.set_title('Bessel Functions of the First Kind')
ax.set_xlabel('$x$')
ax.set_ylabel('$J_n(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 2. Spherical Bessel Equation

The spherical Bessel equation of order $l$ is:

$$x^2 y'' + 2x y' + [x^2 - l(l+1)] y = 0$$

Solutions are spherical Bessel functions $j_l(x) = \sqrt{\pi/(2x)} J_{l+1/2}(x)$.

**Applications:** Scattering theory, partial wave analysis, quantum mechanical hydrogen atom, electromagnetic multipole radiation.

```python
# Symbolic solution
eq = registry.get('spherical_bessel_equation')
print("Spherical Bessel Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot j_0, j_1, j_2
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (0.01, 20)
x_eval = np.linspace(0.01, 20, 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
labels = ['$j_0(x)$', '$j_1(x)$', '$j_2(x)$']

# j_0(x) = sin(x)/x, j_1(x) = sin(x)/x^2 - cos(x)/x
initial_conditions = {
    0: [1.0, 0.0],          # j_0(0) = 1
    1: [0.0, 1.0/3.0],      # j_1(0) -> 0
    2: [0.0, 0.0],           # j_2(0) -> 0
}

for l_val in [0, 1, 2]:
    y0 = initial_conditions[l_val]
    result = eq.numerical_solve(x_span, y0, t_eval=x_eval, params={'l': l_val})
    plot_special_function(result, ax=ax, label=labels[l_val], color=colors[l_val])

ax.set_title('Spherical Bessel Functions')
ax.set_xlabel('$x$')
ax.set_ylabel('$j_l(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 3. Legendre Equation

The Legendre equation of order $n$ is:

$$(1 - x^2) y'' - 2x y' + n(n+1) y = 0$$

Solutions are Legendre polynomials $P_n(x)$ for integer $n$.

**Applications:** Electrostatics (multipole expansion), quantum angular momentum, spherical harmonic decomposition, numerical quadrature.

```python
# Symbolic solution
eq = registry.get('legendre_equation')
print("Legendre Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot P_0 through P_4
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (-0.99, 0.99)
x_eval = np.linspace(-0.99, 0.99, 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

for n_val in range(5):
    # P_n(1) = 1 for all n, derivative from recurrence
    # Start near x=0 and use known values there
    if n_val == 0:
        y0 = [1.0, 0.0]  # P_0(x) = 1
    elif n_val == 1:
        y0 = [0.0, 1.0]  # P_1(x) = x
    elif n_val == 2:
        y0 = [-0.5, 0.0]  # P_2(0) = -1/2
    elif n_val == 3:
        y0 = [0.0, -1.5]  # P_3(0) = 0, P_3'(0) = -3/2
    elif n_val == 4:
        y0 = [3.0/8.0, 0.0]  # P_4(0) = 3/8

    # Solve from x=0 outward in both directions
    result_pos = eq.numerical_solve((0.0, 0.99), y0, t_eval=np.linspace(0.0, 0.99, 250), params={'n': n_val})
    result_neg = eq.numerical_solve((0.0, -0.99), y0, t_eval=np.linspace(0.0, -0.99, 250), params={'n': n_val})

    x_full = np.concatenate([result_neg.t, result_pos.t])
    y_full = np.concatenate([result_neg.y[0], result_pos.y[0]])
    sort_idx = np.argsort(x_full)

    ax.plot(x_full[sort_idx], y_full[sort_idx], color=colors[n_val],
            label=f'$P_{n_val}(x)$', linewidth=1.5)

ax.set_title('Legendre Polynomials $P_n(x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$P_n(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 4. Associated Legendre Equation

The associated Legendre equation is:

$$(1 - x^2) y'' - 2x y' + \left[n(n+1) - \frac{m^2}{1-x^2}\right] y = 0$$

Solutions are associated Legendre functions $P_n^m(x)$, which form the backbone of spherical harmonics.

**Applications:** Spherical harmonics $Y_l^m(\theta,\phi)$, atomic orbital angular dependence, geomagnetic field modeling.

```python
# Symbolic solution
eq = registry.get('associated_legendre_equation')
print("Associated Legendre Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot P_2^0, P_2^1, P_2^2
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (-0.99, 0.99)
x_eval = np.linspace(-0.99, 0.99, 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
labels = ['$P_2^0(x)$', '$P_2^1(x)$', '$P_2^2(x)$']

# P_2^0(x) = (3x^2-1)/2, P_2^1(x) = -3x*sqrt(1-x^2), P_2^2(x) = 3(1-x^2)
# Solve from x=0
ic_list = [
    {'m': 0, 'y0': [-0.5, 0.0]},    # P_2^0(0)
    {'m': 1, 'y0': [0.0, -3.0]},     # P_2^1(0)
    {'m': 2, 'y0': [3.0, 0.0]},      # P_2^2(0)
]

for i, ic in enumerate(ic_list):
    result_pos = eq.numerical_solve(
        (0.001, 0.99), ic['y0'],
        t_eval=np.linspace(0.001, 0.99, 250),
        params={'n': 2, 'm': ic['m']}
    )
    result_neg = eq.numerical_solve(
        (0.0, -0.99), ic['y0'],
        t_eval=np.linspace(0.0, -0.99, 250),
        params={'n': 2, 'm': ic['m']}
    )
    x_full = np.concatenate([result_neg.t, result_pos.t])
    y_full = np.concatenate([result_neg.y[0], result_pos.y[0]])
    sort_idx = np.argsort(x_full)
    ax.plot(x_full[sort_idx], y_full[sort_idx], color=colors[i],
            label=labels[i], linewidth=1.5)

ax.set_title('Associated Legendre Functions $P_2^m(x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$P_2^m(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## 5. Hermite Equation

The Hermite equation is:

$$y'' - 2x y' + 2n y = 0$$

For non-negative integer $n$, solutions are Hermite polynomials $H_n(x)$.

**Applications:** Quantum harmonic oscillator wave functions, Gaussian beam optics, probability theory (Edgeworth series), signal processing.

```python
# Symbolic solution
eq = registry.get('hermite_equation')
print("Hermite Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot H_0 through H_3
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (-4, 4)
x_eval = np.linspace(-4, 4, 500)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# H_0(x) = 1, H_1(x) = 2x, H_2(x) = 4x^2 - 2, H_3(x) = 8x^3 - 12x
# Initial conditions at x=0
hermite_ics = {
    0: [1.0, 0.0],      # H_0(0) = 1, H_0'(0) = 0
    1: [0.0, 2.0],      # H_1(0) = 0, H_1'(0) = 2
    2: [-2.0, 0.0],     # H_2(0) = -2, H_2'(0) = 0
    3: [0.0, -12.0],    # H_3(0) = 0, H_3'(0) = -12
}

for n_val in range(4):
    y0 = hermite_ics[n_val]
    result_pos = eq.numerical_solve((0.0, 4.0), y0,
        t_eval=np.linspace(0.0, 4.0, 250), params={'n': n_val})
    result_neg = eq.numerical_solve((0.0, -4.0), y0,
        t_eval=np.linspace(0.0, -4.0, 250), params={'n': n_val})

    x_full = np.concatenate([result_neg.t, result_pos.t])
    y_full = np.concatenate([result_neg.y[0], result_pos.y[0]])
    sort_idx = np.argsort(x_full)

    ax.plot(x_full[sort_idx], y_full[sort_idx], color=colors[n_val],
            label=f'$H_{n_val}(x)$', linewidth=1.5)

ax.set_title('Hermite Polynomials $H_n(x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$H_n(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
ax.set_xlim(-4, 4)
plt.tight_layout()
plt.show()
```

$$x y'' + (1 - x) y' + n y = 0$$

```python
# Symbolic solution
eq = registry.get('laguerre_equation')
print("Laguerre Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

**Applications:** Numerical approximation theory, polynomial interpolation, filter design, spectral methods for PDEs.

```python
# Symbolic solution
eq = registry.get('chebyshev_first_equation')
print("Chebyshev Equation (First Kind):")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

$$y'' - x y = 0$$

The two linearly independent solutions are the Airy functions $\mathrm{Ai}(x)$ and $\mathrm{Bi}(x)$.

**Applications:** Quantum mechanics near a turning point (linear potential), optics (caustics, rainbow), asymptotic analysis, combinatorics.

```python
# Symbolic solution
eq = registry.get('airy_equation')
print("Airy Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot both Ai(x) and Bi(x)
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (-15, 5)
x_eval = np.linspace(-15, 5, 800)

# Ai(0) = 1/(3^(2/3) * Gamma(2/3)) ~ 0.3550
# Ai'(0) = -1/(3^(1/3) * Gamma(1/3)) ~ -0.2588
ai_ic = [0.355028053887817, -0.258819403792807]
result_ai = eq.numerical_solve(x_span, ai_ic, t_eval=x_eval, params={})
plot_special_function(result_ai, ax=ax, label='$\mathrm{Ai}(x)$', color='#1f77b4')

# Bi(0) = 1/(3^(1/6) * Gamma(2/3)) ~ 0.6149
# Bi'(0) = 3^(1/6) / Gamma(1/3) ~ 0.4483
bi_ic = [0.614926627446001, 0.448288357353826]
result_bi = eq.numerical_solve(x_span, bi_ic, t_eval=x_eval, params={})
plot_special_function(result_bi, ax=ax, label='$\mathrm{Bi}(x)$', color='#ff7f0e')

ax.set_title('Airy Functions $\mathrm{Ai}(x)$ and $\mathrm{Bi}(x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y(x)$')
ax.legend()
ax.grid(True, alpha=0.3)
ax.axhline(0, color='black', linewidth=0.5)
ax.set_ylim(-0.6, 1.5)
plt.tight_layout()
plt.show()
```

## 11. Hypergeometric Equation

The Gauss hypergeometric equation is:

$$x(1-x) y'' + [c - (a+b+1)x] y' - ab\, y = 0$$

Solutions are expressed via the hypergeometric function $_2F_1(a,b;c;x)$.

**Applications:** Nearly all classical special functions are special cases, integrals in mathematical physics, conformal mapping, combinatorics.

```python
# Symbolic solution
eq = registry.get('hypergeometric_equation')
print("Hypergeometric Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot _2F_1 for several (a,b,c) choices
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (0.01, 0.95)
x_eval = np.linspace(0.01, 0.95, 500)
params_list = [
    {'a': 1, 'b': 1, 'c': 2, 'label': '$_2F_1(1,1;2;x) = -\ln(1-x)/x$', 'color': '#1f77b4'},
    {'a': 0.5, 'b': 0.5, 'c': 1.0, 'label': '$_2F_1(1/2,1/2;1;x)$', 'color': '#ff7f0e'},
    {'a': 2, 'b': 1, 'c': 1, 'label': '$_2F_1(2,1;1;x)$', 'color': '#2ca02c'},
]

for p in params_list:
    label = p.pop('label')
    color = p.pop('color')
    # _2F_1(a,b;c;0) = 1
    y0 = [1.0, (p['a'] * p['b']) / p['c']]
    result = eq.numerical_solve(x_span, y0, t_eval=x_eval, params=p)
    plot_special_function(result, ax=ax, label=label, color=color)

ax.set_title('Hypergeometric Functions $_2F_1(a,b;c;x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$_2F_1$')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 12. Confluent Hypergeometric Equation

The confluent hypergeometric (Kummer) equation is:

$$x y'' + (c - x) y' - a\, y = 0$$

Solutions include Kummer's function $M(a,c,x) = {}_1F_1(a;c;x)$ and Tricomi's function $U(a,c,x)$.

**Applications:** Coulomb wave functions, incomplete gamma function, error function, parabolic cylinder functions, hydrogen atom.

```python
# Symbolic solution
eq = registry.get('confluent_hypergeometric_equation')
print("Confluent Hypergeometric Equation:")
print(f"  Form: {eq.description}")
sol = eq.symbolic_solution()
print(f"  General solution: {sol}")
```

```python
# Numerical solution: plot _1F_1 for several parameter choices
fig, ax = plt.subplots(figsize=(10, 6))
x_span = (0.01, 10)
x_eval = np.linspace(0.01, 10, 500)
params_list = [
    {'a': 1, 'c': 1, 'label': '$_1F_1(1;1;x) = e^x$', 'color': '#1f77b4'},
    {'a': 0.5, 'c': 1.0, 'label': '$_1F_1(1/2;1;x)$', 'color': '#ff7f0e'},
    {'a': 2, 'c': 3, 'label': '$_1F_1(2;3;x)$', 'color': '#2ca02c'},
]

for p in params_list:
    label = p.pop('label')
    color = p.pop('color')
    # _1F_1(a;c;0) = 1, _1F_1'(a;c;0) = a/c
    y0 = [1.0, p['a'] / p['c']]
    result = eq.numerical_solve(x_span, y0, t_eval=x_eval, params=p)
    plot_special_function(result, ax=ax, label=label, color=color)

ax.set_title('Confluent Hypergeometric Functions $_1F_1(a;c;x)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$_1F_1$')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
